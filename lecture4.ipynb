# In-Class demonstration

## First Part

- Today we will learn how to use Regression model to predict a stock price.

```python
import pandas as pd

# Use a public AAPL dataset or provide CSV URL
url = 'https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv'
df = pd.read_csv(url, parse_dates=['Date'])
df = df[['Date', 'Total Equity']]
df.rename(columns = {'Total Equity': 'Close'}, inplace=True)
df.dropna(inplace=True)
df.sort_values('Date', inplace=True)
```

```python
df.head(10)
```

```python
df['Close_t-1'] = df['Close'].shift(1)
df['Return'] = df['Close'].pct_change()
df['7D_MA'] = df['Close'].rolling(window=7).mean()
df['30_Vol'] = df['Return'].rolling(window=30).std()
df.dropna(inplace=True)
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df[['Close_t-1', '7D_MA', '30_Vol']]
y = df['Close']


X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size = 0.2)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

y_pred = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'RMSE: {rmse}')
```
